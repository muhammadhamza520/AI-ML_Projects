```python
# %% [markdown]
# ## GRU for Weather Forecasting
# 
# This notebook implements a GRU (Gated Recurrent Unit) model for weather forecasting using historical hourly weather data.

# %% [code]
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.optimizers import Adam

# Set random seed for reproducibility
np.random.seed(42)

# %% [code]
# Load the dataset (assuming files are uploaded to Kaggle)
temperature_df = pd.read_csv('/kaggle/input/historical-hourly-weather-data/temperature.csv')
humidity_df = pd.read_csv('/kaggle/input/historical-hourly-weather-data/humidity.csv')
pressure_df = pd.read_csv('/kaggle/input/historical-hourly-weather-data/pressure.csv')

# Display basic info
print("Temperature DataFrame Shape:", temperature_df.shape)
print("Humidity DataFrame Shape:", humidity_df.shape)
print("Pressure DataFrame Shape:", pressure_df.shape)

# %% [code]
# Data Preprocessing
def preprocess_data(df, city='Vancouver'):
    """Preprocess data for a specific city"""
    # Select city data
    city_data = df[['datetime', city]].copy()
    city_data.columns = ['datetime', 'value']
    
    # Convert datetime and set as index
    city_data['datetime'] = pd.to_datetime(city_data['datetime'])
    city_data.set_index('datetime', inplace=True)
    
    # Handle missing values (forward fill then backfill if needed)
    city_data['value'].ffill(inplace=True)
    city_data['value'].bfill(inplace=True)
    
    return city_data

# Preprocess temperature data for Vancouver
temp_data = preprocess_data(temperature_df, 'Vancouver')
humidity_data = preprocess_data(humidity_df, 'Vancouver')
pressure_data = preprocess_data(pressure_df, 'Vancouver')

# Combine all features
combined_data = pd.concat([
    temp_data.rename(columns={'value': 'temperature'}),
    humidity_data.rename(columns={'value': 'humidity'}),
    pressure_data.rename(columns={'value': 'pressure'})
], axis=1)

# Display sample data
print("\nCombined Data Sample:")
print(combined_data.head())

# %% [code]
# Visualize the data
plt.figure(figsize=(15, 8))
plt.subplot(3, 1, 1)
plt.plot(combined_data['temperature'], label='Temperature')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(combined_data['humidity'], label='Humidity')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(combined_data['pressure'], label='Pressure')
plt.legend()

plt.tight_layout()
plt.show()

# %% [code]
# Prepare data for time series forecasting
def create_sequences(data, n_steps):
    """Convert time series into supervised learning format"""
    X, y = [], []
    for i in range(len(data) - n_steps):
        X.append(data[i:i+n_steps])
        y.append(data[i+n_steps])
    return np.array(X), np.array(y)

# Normalize features
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(combined_data)

# Set parameters
n_steps = 24  # Use 24 hours (1 day) to predict next hour
n_features = scaled_data.shape[1]  # Number of features (temperature, humidity, pressure)

# Create sequences
X, y = create_sequences(scaled_data, n_steps)

# Split into train and test sets (80-20 split)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")

# %% [code]
# Build GRU model
model = Sequential([
    GRU(64, activation='relu', input_shape=(n_steps, n_features), return_sequences=True),
    Dropout(0.2),
    GRU(32, activation='relu'),
    Dropout(0.2),
    Dense(n_features)  # Predict all three features
])

# Compile the model
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])

# Display model summary
model.summary()

# %% [code]
# Define callbacks
callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ModelCheckpoint('best_model.h5', monitor='val_loss', save_best_only=True)
]

# Train the model
history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=32,
    validation_data=(X_test, y_test),
    callbacks=callbacks,
    verbose=1
)

# %% [code]
# Plot training history
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['mae'], label='Train MAE')
plt.plot(history.history['val_mae'], label='Validation MAE')
plt.title('Model MAE')
plt.legend()

plt.tight_layout()
plt.show()

# %% [code]
# Evaluate the model
test_loss, test_mae = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Loss (MSE): {test_loss:.4f}")
print(f"Test MAE: {test_mae:.4f}")

# %% [code]
# Make predictions
y_pred = model.predict(X_test)

# Inverse transform the scaled data
y_test_inv = scaler.inverse_transform(y_test)
y_pred_inv = scaler.inverse_transform(y_pred)

# Create DataFrames for comparison
results_df = pd.DataFrame({
    'Actual Temp': y_test_inv[:, 0],
    'Predicted Temp': y_pred_inv[:, 0],
    'Actual Humidity': y_test_inv[:, 1],
    'Predicted Humidity': y_pred_inv[:, 1],
    'Actual Pressure': y_test_inv[:, 2],
    'Predicted Pressure': y_pred_inv[:, 2]
})

# Display sample predictions
print("\nSample Predictions vs Actuals:")
print(results_df.head(10))

# %% [code]
# Plot actual vs predicted for temperature
plt.figure(figsize=(15, 6))
plt.plot(results_df['Actual Temp'][:100], label='Actual Temperature', color='blue', alpha=0.7)
plt.plot(results_df['Predicted Temp'][:100], label='Predicted Temperature', color='red', alpha=0.5)
plt.title('Temperature: Actual vs Predicted (First 100 Samples)')
plt.xlabel('Time Steps')
plt.ylabel('Temperature')
plt.legend()
plt.show()

# %% [code]
# Save the model for future use
model.save('weather_forecasting_gru_model.h5')
print("Model saved successfully!")
```
